```python
import matplotlib.pyplot as plt
import numpy as np

# Sort the DataFrame by the 'Feature' column
# merged_weights_df = merged_weights_df.sort_values('Feature')

# Number of bars (features) to display
num_bars = 5

# Create an array with the positions of each bar
bar_positions = np.arange(num_bars)

# Set the width of each bar
bar_width = 0.25

# Create the figure
plt.figure(figsize=(12, 6))

# Plot the bars for each model, adjusting the position for each one
plt.bar(bar_positions - bar_width, data["Global Surrogate Model"][:num_bars], color='green', label='Global Surrogate Model', width=bar_width)
plt.bar(bar_positions, data["LIME"][:num_bars], color='red', label='LIME', width=bar_width)
plt.bar(bar_positions + bar_width, data["SHAP"][:num_bars], color='blue', label='SHAP', width=bar_width)

# Set the y-ticks to be the feature names
plt.xticks(bar_positions, data["Feature"][:num_bars])

plt.ylabel('Feature Importance', size = 15)
plt.xlabel('Feature', size = 15)
plt.legend(fontsize = 15)
# change the names of the x ticks to 1, 2, 3, 4, 5
plt.xticks([0, 1, 2, 3, 4], ['Debt to Income \n Ratio missing', 'Interest Rate', 'Debt to Income \n Ratio >60%', 'County Code', 'Loan Type VA'], size = 15, rotation=30, ha='right')
# plt.xticks(rotation=30, ha='right', size = 15)
plt.yticks(size = 15)
plt.show()
```
